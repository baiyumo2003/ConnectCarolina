```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.expected_conditions import presence_of_element_located
import time
import _thread

onyen='' #Your onyen 
password='' #your password
Subject='math'  #subject you need like COMP MATH BIOL 
Subject=Subject.upper()
classNumber='528'   #COMP 210 course number
sectionNumber='3430' # which Section number you would love to enroll
driver = webdriver.Chrome()
driver.get("https://pa.cc.unc.edu/psp/paprd/EMPLOYEE/SA/c/SA_LEARNER_SERVICES.SSS_STUDENT_CENTER.GBL")
driver.find_element(By.XPATH,"//*[@id='username']").send_keys(onyen)
driver.find_element(By.XPATH,"//*[@id='nextBtn']").click()
while(True):
    try:
        driver.find_element(By.XPATH,"//*[@id='password']").send_keys(password)
        break
    except:
        pass
driver.find_element(By.XPATH,"//*[@id='submitBtn']").click()

def monitoringStatus():
    while(True):
        # driver.find_element(By.XPATH,"//*[contains(text(),'Search')]").click()
        driver.switch_to.frame('ptifrmtgtframe')
        driver.find_element(By.XPATH,"//*[@id='DERIVED_SSS_SCR_SSS_LINK_ANCHOR1']").click()
        while(True):
            try:
                driver.find_element(By.XPATH,"//*[@id='SSR_CLSRCH_WRK_SUBJECT$0']").send_keys(Subject)
                break
            except:
                pass

        driver.find_element(By.XPATH,"//*[@id='SSR_CLSRCH_WRK_SSR_OPEN_ONLY$3']").click()
        driver.find_element(By.XPATH,"//*[@id='SSR_CLSRCH_WRK_CATALOG_NBR$1']").send_keys(classNumber)
        driver.find_element(By.XPATH,"//*[@id='CLASS_SRCH_WRK2_SSR_PB_CLASS_SRCH']").click()
        while(True):
            try:
                driver.find_element(By.LINK_TEXT,sectionNumber).click()
                break
            except:
                pass

        while(True):
            try:
                classStatus=driver.find_element(By.XPATH,"//*[@id='SSR_CLS_DTL_WRK_SSR_DESCRSHORT']").text
                break
            except:
                pass
        if(classStatus!='Closed'):
            sendEmailForOpen()
            addCourse()
            break
        driver.refresh()

def addCourse():
    
    driver.find_element(By.XPATH,"//*[@id='PSTAB']/table/tbody/tr/td[2]/a/span").click()
    while(True):
        try:
            driver.find_element(By.XPATH,"//*[@id='win0divDERIVED_SSTSNAV_SSTS_NAV_SUBTABS']/div/table/tbody/tr/td[8]/a/span").click()
            break
        except:
            pass
    xPath="//*[@id='P_SELECT$0']"
    xPath=xPath.replace('0',str(locateAdding(classNumber,sectionNumber)))
    driver.find_element(By.XPATH,xPath).click()
    driver.find_element(By.XPATH,"//*[@id='DERIVED_REGFRM1_LINK_ADD_ENRL']").click()
    while(True):
        try:
            driver.find_element(By.XPATH,"//*[@id='DERIVED_REGFRM1_SSR_PB_SUBMIT']").click()
            break
        except:
            pass
    sendEmail()
    

def locateAdding(classNumber,sectionNumber ):
    startIndex="//*[@id='P_CLASS_NAME$0']"
    startIndex=list(startIndex)
    a=0
    while(True):
        startIndex[22]=str(a)
        tempText=driver.find_element(By.XPATH,''.join(startIndex)).text
        if(classNumber in tempText and sectionNumber in tempText):
            return a
        a+=1

def sendEmail():
    from email.message import EmailMessage
    import ssl
    import smtplib

    email_sender='' #given gmail
    email_password='' #password generated
    email_receiver='' #the email that you would like to send

    subject='success for joining'+Subject+" "+classNumber

    body="The "+Subject+" "+classNumber+" is successfully joined"
    em=EmailMessage()
    em['From']=email_sender
    em['To']=email_receiver
    em['Subject']=subject
    em.set_content(body)

    context=ssl.create_default_context()

    with smtplib.SMTP_SSL('smtp.gmail.com',465,context=context) as smtp:
        smtp.login(email_sender,email_password)
        smtp.sendmail(email_sender,email_receiver,em.as_string())

def sendEmailForOpen():
    from email.message import EmailMessage
    import ssl
    import smtplib

    email_sender='' #given gmail
    email_password='' #password generated
    email_receiver='' #the email that you would like to send

    subject='Open for '+Subject+" "+classNumber
    body="The "+Subject+" "+classNumber+" is now Open"
    em=EmailMessage()
    em['From']=email_sender
    em['To']=email_receiver
    em['Subject']=subject
    em.set_content(body)

    context=ssl.create_default_context()

    with smtplib.SMTP_SSL('smtp.gmail.com',465,context=context) as smtp:
        smtp.login(email_sender,email_password)
        smtp.sendmail(email_sender,email_receiver,em.as_string())

sendEmailForOpen()     
monitoringStatus()
```
